# Statistics Part-2

1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population parameter based on sample data. It involves formulating two competing hypotheses (null and alternative), collecting data, and using statistical analysis to determine which hypothesis is more strongly supported by the evidence.

2. What is the null hypothesis, and how it differ from the alternative hypothesis?
- THe null hypothesis is a foundational concept in statistical hypothesis testing. It is a statement that there is no effect, no difference, or no relationship between variables. It represents the default or status quo assumption that researchers aim to test.

- The alternative hypothesis is a statement that contradicts the nul hypothesis. it proposes that there is an effect, a difference, or a relationship between variables.

3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the probability threshold for rejecting the null hypothesis, typically set at 0.05 (5%). It represents the maximum acceptable probability of making a Type I error (false positive). It's important because it determines how strong the evidence must be to reject H₀.

4. What does a P-value represents in hypothesis testing?
- A P-value represents the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true. It measures the strength of evidence against H₀.

5. How do you interpret the P-value in hypothesis testing?
- P-value ≤ α: Reject H₀ (results are statistically significant)
- P-value > α: Fail to reject H₀ (results are not statistically significant)
A smaller P-value indicates stronger evidence against H₀.

6. What are Type 1 and type 2 errors in hypothesis testing?
- Type I error (α)*: False positive - rejecting H₀ when it's actually true
- Type II error (β)*: False negative - failing to reject H₀ when it's actually false
- Power (1-β)*: Probability of correctly rejecting a false H₀

7. What is the difference between a one-tailed and two-tailed test in hypothesis testing?
- One-tailed test*: Tests for an effect in one direction only (e.g., "greater than" or "less than")
- Two-tailed test*: Tests for an effect in both directions (e.g., "not equal to")

8. What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is a statistical test that uses the standard normal distribution (Z-distribution) to compare sample means to population means when:
1. Population parameters (mean and standard deviation) are known
2. Sample size is large (typically n > 30)
3. Data is normally distributed

9. How dp you calculate the Z-score, and what does it represent in hypothesis testing?
- Z = (X̄ - μ) / (σ/√n)
Where:
- X̄ = sample mean
- μ = population mean
- σ = population standard deviation
- n = sample size

The Z-score represents how many standard deviations the sample mean is from the population mean.

10. What is the T-distribution , and when should it be used instead of the normal distribution?
- The T-distribution is a probability distribution similar to the normal distribution but with heavier tails. It should be used when:
 * Sample size is small (typically n < 30)
 * Population standard deviation is unknown
 * Data are approximately normally distributed

11. What is the difference between a Z-test and a T-test?
- Z-test: Uses normal distribution, requires known population parameters, used for large samples
- T-test: Uses t-distribution, doesn't require known population standard deviation, used for small samples

12. What is the T-test, and how does it used in hypothesis teating?
- A T-test is a statistical test that compares means using the t-distribution. Common types:
 * One-sample t-test*: Compares sample mean to a known value
 * Independent samples t-test: Compares means of two independent groups
 * Paired t-test*: Compares means of related measurements (before/after)

13. What is the relationship bteween Z-test and T-test in hypothesis testing?
- Both are used to test hypotheses about means. As sample size increases, the t-distribution approaches the normal distribution, making t-test results similar to z-test results. For large samples (n > 30), they often yield similar conclusions.

14. What is a confidence interval, and how is it used to interpret statical results?
- A confidence interval is a range of values likely to contain a population parameter with a specified probability (e.g., 95% CI). Interpretation: "We are 95% confident the true population parameter lies within this interval."

15. What is the margin of error, and how does it used to interpret statistical results?
- Margin of error is half the width of the confidence interval, representing the maximum expected difference between the sample statistic and population parameter. A larger margin of error creates a wider confidence interval, indicating less precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem updates the probability of a hypothesis as new evidence is acquired:
P(A|B) = [P(B|A) × P(A)] / P(B)

Significance:
- Provides a mathematical framework for updating beliefs
- Foundation of Bayesian statistics
- Used in machine learning, medical testing, and many other fields

17. What is the Chi-square distribution, and when is it used?
- The Chi-square (χ²) distribution is a right-skewed distribution used for:
  * Tests of independence in contingency tables
  * Goodness-of-fit tests
  * Variance tests for normally distributed data

18. WHat is the CHi-square goodness of fit test, and how is it applied?
- A test that determines whether sample data match a theoretical distribution. Steps:
 1. Specify expected distribution
 2. Calculate χ² = Σ[(Observed - Expected)²/Expected]
 3. Compare to χ² distribution with appropriate df

19. WHat is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is a right-skewed distribution used for:
  * Comparing two variances (F-test)
  * ANOVA (analysis of variance)
  * Regression analysis

20. What is an ANOVA tesy, and what are its assumptions?
- ANOVA (Analysis of Variance) tests differences among group means. Assumptions:
 1. Independence of observations
 2. Normally distributed populations
 3. Homogeneity of variances (equal variances across groups)

21. What are the different types of ANOVA tests?
- *One-way ANOVA*: Compares means across one independent variable with ≥3 levels
- Two-way ANOVA: Examines effect of two independent variables
- Repeated measures ANOVA: For dependent samples measured multiple times
- Factorial ANOVA: For multiple independent variables with interactions

22. WHat is the F-test, and how does it relate to hypothesis testing?
- The F-test compares variances by calculating the ratio of two variances (F = s₁²/s₂²). In hypothesis testing, it's used to:
  * Test equality of two variances
  * Assess overall significance in ANOVA and regression
  * Compare nested statistical models

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = np.random.normal(loc=105, scale=15, size=30)  # Simulated data with mean 105

# Population parameters
pop_mean = 100
pop_std = 15

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-score
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

# Two-tailed test at 0.05 significance level
critical_value = norm.ppf(0.975)  # 1.96 for alpha=0.05
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"Critical value: ±{critical_value:.2f}")
print(f"P-value: {p_value:.4f}")

if abs(z_stat) > critical_value:
    print("Reject null hypothesis (significant difference)")
else:
    print("Fail to reject null hypothesis (no significant difference)")


## 16. Create a conf
```

16. Create a confidence interval for a dataset using Python and interpet the result.

```python
import numpy as np
from scipy.stats import norm

# Generate sample data
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate 95% confidence interval
confidence_level = 0.95
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

margin_of_error = norm.ppf((1 + confidence_level) / 2) * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean lies between",
      f"{ci_lower:.2f} and {ci_upper:.2f}")
```

24. Write a Python function to calculate the sekewness and kurtosis of a dataset fenerated from a normal disrtibution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, norm

def calculate_skewness_kurtosis(sample_size=1000):
    """
    Calculate and display skewness and kurtosis of a normally distributed dataset.

    Parameters:
    sample_size (int): Number of samples to generate from normal distribution

    Returns:
    tuple: (skewness, kurtosis) of the dataset
    """
    # Generate normally distributed data
    mu, sigma = 0, 1  # mean and standard deviation
    data = np.random.normal(mu, sigma, sample_size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data, fisher=True)  # Fisher's definition (normal = 0)

    # Print results
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis: {data_kurtosis:.4f}")

    # Plot the distribution
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Sample Data')

    # Plot the theoretical normal PDF for comparison
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, sigma)
    plt.plot(x, p, 'k', linewidth=2, label='Theoretical Normal')

    title = (f'Normal Distribution\n'
             f'Skewness: {data_skewness:.4f}, Kurtosis: {data_kurtosis:.4f}')
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    return data_skewness, data_kurtosis

# Example usage
skewness, kurt = calculate_skewness_kurtosis(1000)
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the result.

```python
import numpy as np
from scipy import stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """Perform one-sample Z-test"""
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std/np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject null hypothesis (significant difference)")
    else:
        print("Fail to reject null hypothesis (no significant difference)")

    return z_score, p_value

# Example usage
pop_mean, pop_std = 100, 15
sample = np.random.normal(loc=105, scale=15, size=30)
z, p = z_test(sample, pop_mean, pop_std)
```

```python
import numpy as np
from scipy import stats

def one_sample_ztest(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z_score = (sample_mean - pop_mean) / se

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_score)
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_score)

    print(f"Sample mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject H0 ({alternative} alternative)")
    else:
        print("Fail to reject H0")

    return z_score, p_value

# Example
data = np.random.normal(loc=102, scale=15, size=50)
one_sample_ztest(data, pop_mean=100, pop_std=15)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate data
pop_mean, pop_std = 50, 10
sample = np.random.normal(loc=53, scale=10, size=30)

# Perform test
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - pop_mean) / (pop_std/np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Visualization
x = np.linspace(pop_mean - 4*pop_std/np.sqrt(n), pop_mean + 4*pop_std/np.sqrt(n), 1000)
y = stats.norm.pdf(x, loc=pop_mean, scale=pop_std/np.sqrt(n))

plt.plot(x, y, label='Null Distribution')
plt.axvline(pop_mean, color='black', linestyle='--')
plt.axvline(sample_mean, color='red', label='Sample Mean')

# Critical regions
alpha = 0.05
z_crit = stats.norm.ppf(1-alpha/2)
lower = pop_mean - z_crit * pop_std/np.sqrt(n)
upper = pop_mean + z_crit * pop_std/np.sqrt(n)
plt.axvspan(x.min(), lower, color='red', alpha=0.2, label='Rejection Region')
plt.axvspan(upper, x.max(), color='red', alpha=0.2)

plt.title(f"Two-tailed Z-test (p = {p_value:.4f})")
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

5. Create Python function that calculates and visualizes type 1 and type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_errors(true_mean=0, effect_size=0.5, n=30, alpha=0.05):
    """Visualize Type I and Type II errors"""
    std = 1
    se = std/np.sqrt(n)

    # Null and alternative distributions
    x = np.linspace(-3, 3 + effect_size, 1000)
    null_dist = stats.norm.pdf(x, loc=0, scale=se)
    alt_dist = stats.norm.pdf(x, loc=effect_size, scale=se)

    # Critical value
    z_crit = stats.norm.ppf(1 - alpha/2)
    crit_lower = -z_crit * se
    crit_upper = z_crit * se

    # Type II error (beta)
    beta = stats.norm.cdf(crit_upper, loc=effect_size, scale=se) - \
           stats.norm.cdf(crit_lower, loc=effect_size, scale=se)
    power = 1 - beta

    # Plot
    plt.plot(x, null_dist, label='Null (H0)')
    plt.plot(x, alt_dist, label='Alternative (H1)')

    # Type I error (alpha)
    plt.fill_between(x, null_dist, where=(x <= crit_lower)|(x >= crit_upper),
                    color='red', alpha=0.3, label=f'Type I Error (α={alpha})')

    # Type II error (beta)
    plt.fill_between(x, alt_dist, where=(x > crit_lower)&(x < crit_upper),
                    color='blue', alpha=0.3, label=f'Type II Error (β={beta:.2f})')

    plt.axvline(crit_lower, color='black', linestyle=':')
    plt.axvline(crit_upper, color='black', linestyle=':')
    plt.title(f"Power = {power:.2f}")
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

plot_errors(effect_size=0.8)
```

6. Write a Python program to perform an independemt T-test and interpret the result.

```python
import numpy as np
from scipy import stats

def independent_ttest(group1, group2, alpha=0.05):
    """Perform independent samples t-test"""
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)

    print(f"Group 1 Mean: {np.mean(group1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject null hypothesis - significant difference")
    else:
        print("Fail to reject null hypothesis - no significant difference")

    return t_stat, p_value

# Example
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
independent_ttest(group1, group2)
```

9. Write a Python function to calculate the confidence interval for a sample mean and example its significance.

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (array-like): Sample data
    confidence_level (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
    tuple: (lower_bound, upper_bound)
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Get critical value (z for large samples, t for small samples)
    if sample_size >= 30:
        critical_value = stats.norm.ppf((1 + confidence_level) / 2)
    else:
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1)

    # Calculate margin of error
    margin_of_error = critical_value * standard_error

    # Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage
sample_data = np.random.normal(loc=50, scale=10, size=100)  # Normally distributed data with mean 50
ci = calculate_confidence_interval(sample_data, 0.95)
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

```

11. Implement a Bayesiap inference method using Bayes' theorem in Python and explain the process.

```python
def bayesian_inference(prior, likelihood, evidence):
    """
    Perform Bayesian inference using Bayes' Theorem.

    Parameters:
    prior (float): Prior probability P(A)
    likelihood (float): Likelihood P(B|A)
    evidence (float): Marginal probability P(B)

    Returns:
    float: Posterior probability P(A|B)
    """
    posterior = (prior * likelihood) / evidence
    return posterior

# Example: Disease testing scenario
# Suppose:
# - Prevalence of disease (prior): 1% or 0.01
# - Test sensitivity (likelihood P(Test+|Disease+)): 99% or 0.99
# - Test specificity (P(Test-|Disease-)): 95% or 0.95
# - What's P(Disease+|Test+)?

prior = 0.01  # P(Disease+)
sensitivity = 0.99  # P(Test+|Disease+)
specificity = 0.95  # P(Test-|Disease-)

# Calculate P(Test+) = P(Test+|Disease+)*P(Disease+) + P(Test+|Disease-)*P(Disease-)
p_test_positive = (sensitivity * prior) + ((1 - specificity) * (1 - prior))

posterior = bayesian_inference(prior, sensitivity, p_test_positive)
print(f"Probability of having the disease given a positive test result: {posterior:.4f} or {posterior*100:.2f}%")
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
from scipy.stats import chi2_contingency

def chi_square_test(observed):
    """
    Perform Chi-square test for independence between two categorical variables.

    Parameters:
    observed (array-like): Contingency table of observed frequencies

    Returns:
    tuple: (chi2_stat, p_value, dof, expected)
    """
    chi2_stat, p_value, dof, expected = chi2_contingency(observed)
    return chi2_stat, p_value, dof, expected

# Example: Test if gender is independent of preference for a new product
# Contingency table:
#         Like Dislike
# Male     50    30
# Female   45    25
observed = np.array([[50, 30], [45, 25]])

chi2_stat, p_value, dof, expected = chi_square_test(observed)
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print("Expected frequencies:")
print(expected)

if p_value < 0.05:
    print("Reject null hypothesis: Variables are dependent")
else:
    print("Fail to reject null hypothesis: Variables may be independent")

```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed (array-like): Contingency table of observed frequencies

    Returns:
    numpy.ndarray: Table of expected frequencies
    """
    # Calculate row and column totals
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()

    # Calculate expected frequencies
    expected = np.outer(row_totals, col_totals) / grand_total

    return expected

# Example usage
observed = np.array([[50, 30], [45, 25]])
expected = calculate_expected_frequencies(observed)
print("Observed frequencies:")
print(observed)
print("\nExpected frequencies:")
print(expected)
```
